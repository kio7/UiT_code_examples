```python
actual_weight = 78.56
zns = [78.2, 78.9, 78.1, 78.3, 78.4, 78.9, 78.6, 78.8, 78.7, 78.4, 79, 78.2, 78.7, 78.7, 78.3]
estimates = [zns[0]]

def new_estimate(new_pos, n):
    return estimates[-1] + 1/n*(new_pos - estimates[-1])
```

```python
for i in range(1, len(zns)):
    estimates.append(new_estimate(zns[i], i))
    
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(zns, label = 'Measurements')
plt.plot(estimates, label = 'Estimates')
plt.axhline(actual_weight, color = 'green' ,label = 'Actual weight')
plt.legend()
plt.xlabel('Attempt number')
plt.ylabel('Weight')
```

```python
lis = []

with open("noisy_position.txt", "r") as file:
    file.readline() # First line is name of measurements and not actual measurements.
    for line in file:
        lis.append(line.strip().split("    "))

zns = []
for elm in lis:
    zns.append(float(elm[1]))

estimated_v = [0]
estimated_pos = [zns[0]]
alpha = 0.12
beta = 0.05

for zn in zns[1:]:
    e_v = estimated_v[-1] + beta*((zn - estimated_v[-1])/0.5)
    e_pos = estimated_pos[-1] + alpha*(zn - estimated_pos[-1])
    estimated_v.append(e_v)
    estimated_pos.append(e_pos)
```

```python
plt.plot(zns, label = 'Measurements', color = 'red')
plt.plot(estimated_pos, label = 'Estimated position', color = 'blue')
plt.plot(estimated_v, label = 'Estimated velocity', color = 'green')

plt.legend()
plt.xlabel('Time in secounds')
plt.ylabel('Position')
```
